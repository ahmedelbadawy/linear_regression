# Linear Regression Model

```python
from linear_regression import *
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## Univariate Data

```python
X,Y = read_dat("univariateData.dat")
X = feature_normalize(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1)
weights,no_iterations,error = fit_model(X_train,Y_train,100,0.3)
plt.scatter(no_iterations,error,color='red')
plt.xlabel("# Iterations")
plt.ylabel("Cost function")
print("Cost fuction is " , compute_cost(X_test,Y_test,weights))
print("Mean Absolute Error(MAE) is ",evaluate_performance(Y_test,predict(X_test,weights)))
```

```python
plt.scatter(X_train[:,1],Y_train[:,0],color = "green")
plt.plot(X_train[:,1],predict(X_train,weights),color='red')
plt.xlabel("X (normalized)")
plt.ylabel("Y")
```

## Multivariate Data

```python
X,Y = read_dat("multivariateData.dat")
X = feature_normalize(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1)
weights,no_iterations,error = fit_model(X_train,Y_train,1000,0.3)
plt.scatter(no_iterations,error,color='red')
plt.xlabel("# Iterations")
plt.ylabel("Cost function")
print("Cost fuction is " , compute_cost(X_test,Y_test,weights))
print("Mean Absolute Error(MAE) is ",evaluate_performance(Y_test,predict(X_test,weights)))
```
